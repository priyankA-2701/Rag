```python
%%writefile app.py

# Importing necessary libraries
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
import string
import uuid
import random
import hashlib

import os
from PyPDF2 import PdfReader
from langchain.chat_models import ChatOpenAI
from langchain.chains import LLMChain
from langchain import PromptTemplate
from langchain.chains.summarize import load_summarize_chain
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.docstore.document import Document
from langchain.schema import SystemMessage, HumanMessage


from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import re

# ---------------------------------------------------------------------------------------

# Set up OpenAI API key
open_api_key = ""
os.environ["OPENAI_API_KEY"] = open_api_key

# Initialize the OpenAI model
llm = ChatOpenAI(model_name='gpt-3.5-turbo')

# ---------------------------------------------------------------------------------------

# Set app theme, colors, and fonts
st.set_page_config(
    page_title="Text Mining Project - Khan, Manda",
    page_icon=":clipboard:",  # You can use any emoji for the page icon
    layout="wide",
    initial_sidebar_state="expanded",
)

# Add a logo at the top of the sidebar
st.sidebar.image("/content/drive/MyDrive/TM Working/Simplifying/logo.png", use_column_width=True)  # Provide the correct path to your logo

# ---------------------------------------------------------------------------------------

def home_page():
    # Video from a file path
    st.video("/content/drive/MyDrive/TM Working/Simplifying/title.mp4")

# ---------------------------------------------------------------------------------------

def about_page():
    #st.title("About the Project")
    # Video from a file path
    st.video("/content/drive/MyDrive/TM Working/Simplifying/us.mp4")

    st.markdown("""
The SummarAIze  is a versatile application designed to assist users in summarizing text or speeches, extract keywords, create word clouds
and extracting information from PDF documents through natural language interactions.
Leveraging cutting-edge language models and streamlit for the user interface,
this application provides a seamless experience for summarizing speeches, texts , extracting key insights from PDFs,
and answering questions based on the provided documents.​
    """)

    st.markdown("""

    ### Features

    1. **Summarization of Speeches**: Users can input speech texts directly into the chatbot, which then generates a concise summary of the provided text.

    2. **PDF Summarization**: Users can upload PDF documents containing speeches or other textual content. The chatbot extracts text from the PDF, summarizes it, and presents the summarized version to the user.

    3. **Question Answering**: After summarizing a PDF, users can ask questions related to the document's content. The chatbot utilizes the extracted text to provide relevant answers to the user's questions.

    4. **Keyword Extraction and Word Cloud Generation**: Users can input text to extract keywords and generate a word cloud that visually represents the frequency and importance of words in the text.

    ### How to Use

    1. **Text Input**:
       - Select the "Text Input" option.
       - Enter the speech text into the text area.
       - Click on the "Summarize" button to generate the summary.

    2. **Upload PDF**:
       - Choose the "Upload PDF" option.
       - Upload a PDF file containing the speech or document to be summarized.
       - The chatbot will process the PDF, summarize its content, and display the summary.
       - Optionally, input a question related to the PDF content in the provided text field.
       - Click on the "Ask" button to receive an answer based on the provided question.

    3. **Keyword Extraction and Word Cloud**:
       - Select the "Keyword Extraction and Word Cloud" option.
       - Enter the text into the provided text area.
       - Click on the "Extract Keywords and Generate word Cloud" button to display the extracted keywords.

    """)

# ---------------------------------------------------------------------------------------

def insights_page():
    #st.title("Need for the Study")
    st.image("/content/drive/MyDrive/TM Working/Simplifying/NS.JPG", use_column_width=True)
    st.video("/content/drive/MyDrive/TM Working/Simplifying/flowchart.mp4")
    st.markdown("""

    ### Need for the Study of Text Summarization and Analysis Applications

    In the era of information overload, the ability to efficiently process and extract meaningful insights from large volumes of text has become increasingly critical. The demand for applications that can summarize, analyze, and visualize text and documents is driven by several factors:

    **1. Exponential Growth of Digital Content:**

    - **Information Overload:** With the vast amount of information available online and in digital formats, individuals and organizations are overwhelmed with data. Text summarization tools help in condensing lengthy documents into concise summaries, enabling quicker understanding and decision-making.

    - **Document Management:** Businesses, educational institutions, and research organizations generate and handle extensive documentation. Efficient PDF processing and keyword extraction applications facilitate better document management, retrieval, and utilization of information.

    **2. Enhanced Productivity and Efficiency:**

    - **Time Savings:** Manual summarization and information extraction from documents are time-consuming and prone to errors. Automated tools streamline these processes, saving valuable time and reducing the cognitive load on individuals.

    - **Improved Decision-Making:** By providing quick access to summarized information and key insights, these applications enable more informed and timely decision-making across various domains, including business, healthcare, and academia.

    **3. Advances in Natural Language Processing (NLP):**

    - **Technological Innovations:** Recent advancements in NLP, powered by deep learning models like BERT, GPT, and T5, have significantly improved the accuracy and efficacy of text summarization and keyword extraction algorithms. These innovations make it feasible to develop more sophisticated and reliable applications.

    - **AI Integration:** The integration of AI into text processing tools allows for more nuanced and context-aware analysis, enhancing the overall performance and user experience.

    **4. Demand for Data-Driven Insights:**

    - **Business Intelligence:** In today's competitive landscape, businesses rely heavily on data-driven insights for strategic planning and operations. Applications that can analyze text data and present it visually through word clouds and interactive dashboards add immense value.

    - **Research and Academia:** Researchers and educators benefit from tools that can quickly summarize vast amounts of literature, extract relevant keywords, and visualize trends, thereby accelerating the pace of discovery and learning.

    **5. Accessibility and Inclusivity:**

    - **Improving Access to Information:** By summarizing complex documents and extracting key information, these applications make content more accessible to a broader audience, including those with cognitive or reading challenges.

    - **Language Barriers:** Multilingual text processing capabilities help in breaking down language barriers, enabling users from diverse linguistic backgrounds to access and comprehend information effectively.

    """)

# ---------------------------------------------------------------------------------------

def lor():
    #st.title("Libraries Used")
    st.image("/content/drive/MyDrive/TM Working/Simplifying/libraries.jpg", use_column_width=True)

    st.markdown("""
    ### Libraries Used in the Application:

    This application leverages several powerful libraries to provide functionalities such as text summarization, PDF processing, keyword extraction, and data visualization. Below is a brief overview of the libraries used:

    - **Streamlit**: A framework for creating interactive web applications in Python. It allows us to build and deploy custom applications quickly and efficiently.

    - **Pandas**: A data manipulation and analysis library. It provides data structures and functions needed to manipulate structured data seamlessly.

    - **Matplotlib**: A plotting library used for creating static, animated, and interactive visualizations in Python.

    - **Seaborn**: A data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

    - **Plotly**: An interactive graphing library that makes it easy to create interactive plots, dashboards, and data applications.

    - **NumPy**: A library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays.

    - **String**: A module that provides a collection of string operations.

    - **UUID**: A module that provides immutable universally unique identifiers (UUIDs).

    - **Random**: A module that implements pseudo-random number generators for various distributions.

    - **Hashlib**: A module that provides a common interface to many secure hash and message digest algorithms.

    - **OS**: A module that provides a way of using operating system dependent functionality like reading or writing to the file system.

    - **PyPDF2**: A library for reading and manipulating PDF files. It allows for PDF parsing and text extraction.

    - **LangChain**: A framework for developing applications powered by language models. It provides tools for managing chains of language model calls and their configurations.

        - **ChatOpenAI**: A module within LangChain for interacting with OpenAI's language models.

        - **LLMChain**: A module for creating and managing chains of language model calls.

        - **PromptTemplate**: A module for creating templates for prompts used in language model interactions.

        - **load_summarize_chain**: A function for loading pre-configured chains for summarizing text.

        - **RecursiveCharacterTextSplitter**: A module for splitting text into smaller chunks based on character limits.

        - **Document**: A module for creating document objects that can be processed by language models.

        - **SystemMessage, HumanMessage**: Modules for creating messages in a chat format for interactions with language models.

    - **WordCloud**: A library for generating word clouds from text. It visualizes text data where the size of each word indicates its frequency or importance.

    - **Re**: A module that provides support for regular expressions in Python.

    - **Counter**: A module from the collections library that provides a convenient way to count occurrences of elements in a collection.

    These libraries collectively provide the foundation for building a robust and interactive text summarization and analysis application.
    """)

# ---------------------------------------------------------------------------------------

def pros():
    st.title("Strengths and Limitations of Using the Model")
    st.image("/content/drive/MyDrive/TM Working/Simplifying/P&C.JPG", use_column_width=True)
    st.image("/content/drive/MyDrive/TM Working/Simplifying/p&c2.jpg", use_column_width=True)

# ---------------------------------------------------------------------------------------

def conclude():
    st.snow()
    #st.title("Conclusion")
    st.image("/content/drive/MyDrive/TM Working/Simplifying/Conclusion.JPG", use_column_width=True)
    st.image("/content/drive/MyDrive/TM Working/Simplifying/conclusion 2.jpg", use_column_width=True)

# ---------------------------------------------------------------------------------------

def references():
    st.image("/content/drive/MyDrive/TM Working/Simplifying/ref.jpg", use_column_width=True)
    st.markdown("""
    ### References

    - DataCamp. (n.d.). What is an LLM? A guide on large language models. DataCamp. [Retrieved from](https://www.datacamp.com/blog/what-is-an-llm-a-guide-on-large-language-models)

    - Streamlit Team. (2021, May 5). Langchain Tutorial #1: Build an LLM-Powered App in 18 Lines of Code. Streamlit Blog. [Retrieved from](https://blog.streamlit.io/langchain-tutorial-1-build-an-llm-powered-app-in-18-lines-of-code/)

    - Langchain. (n.d.). Use Cases: Chatbots. Langchain. [Retrieved from](https://python.langchain.com/v0.1/docs/use_cases/chatbots/)

    - FreeCodeCamp. (n.d.). Beginner’s Guide to LangChain. FreeCodeCamp. [Retrieved from](https://www.freecodecamp.org/news/beginners-guide-to-langchain/)

    - PyPI. (n.d.). Langchain. [Retrieved from](https://pypi.org/project/langchain/)

    - IBM. (n.d.). Langchain. IBM. [Retrieved from](https://ibm.com/topics/langchain)

    - Real Python. (n.d.). Build a Chatbot: Python Chatbot Tutorial. Real Python. [Retrieved from](https://realpython.com/build-a-chatbot-python-chatterbot/)

    - Analytics Vidhya. (2021, July 20). Build a Simple Chatbot using Python and NLTK. Analytics Vidhya. [Retrieved from](https://www.analyticsvidhya.com/blog/2021/07/build-a-simple-chatbot-using-python-and-nltk/)
    """)


# ---------------------------------------------------------------------------------------

def apikey():
    st.image("/content/drive/MyDrive/TM Working/Simplifying/api 1.jpg", use_column_width=True)
    st.image("/content/drive/MyDrive/TM Working/Simplifying/api 2.jpg", use_column_width=True)
    st.image("/content/drive/MyDrive/TM Working/Simplifying/api 3.jpg", use_column_width=True)

# ---------------------------------------------------------------------------------------

# Function to summarize text
def summarize_text(speech):
    chat_messages = [
        SystemMessage(content='You are an expert assistant with expertise in summarizing speeches'),
        HumanMessage(content=f'Please provide a short and concise summary of the following speech:\n TEXT: {speech}')
    ]
    summary = llm(chat_messages).content
    return summary

# Function to summarize PDF
def summarize_pdf(file):
    pdfreader = PdfReader(file)
    text = ''
    for page in pdfreader.pages:
        content = page.extract_text()
        if content:
            text += content

    # Split text into chunks
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=20)
    chunks = text_splitter.create_documents([text])

    # Summarize using map-reduce
    chain = load_summarize_chain(
        llm,
        chain_type='map_reduce',
        verbose=False
    )
    summary = chain.run(chunks)
    return summary, text

# Function to ask questions about the PDF
def ask_question(text, question):
    chat_messages = [
        SystemMessage(content='You are an expert assistant with expertise in answering questions based on provided documents.'),
        HumanMessage(content=f'Text: {text}\n\nQuestion: {question}')
    ]
    answer = llm(chat_messages).content
    return answer

# Function for the text summarizer page
def text_summarizer_page():
    #st.title("Text Summarizer")
    st.video("/content/drive/MyDrive/TM Working/Simplifying/text.mp4")
    speech = st.text_area("Enter the text you want to summarize:")
    if st.button("Summarize"):
        if speech:
            summary = summarize_text(speech)
            st.write("Summary:")
            st.write(summary)
        else:
            st.write("Please enter the text you want to summarize.")

# Function for the PDF summarizer page
def pdf_summarizer_page():
    #st.title("PDF Summarizer")
    st.video("/content/drive/MyDrive/TM Working/Simplifying/doc.mp4")
    uploaded_file = st.file_uploader("Upload a PDF file", type="pdf")
    if uploaded_file:
        summary, text = summarize_pdf(uploaded_file)
        st.write("Summary:")
        st.write(summary)

        question = st.text_input("Ask a question about the PDF:")
        if st.button("Ask"):
            if question:
                answer = ask_question(text, question)
                st.write("Answer:")
                st.write(answer)
            else:
                st.write("Please enter a question.")

# ---------------------------------------------------------------------------------------

# Function to extract keywords using AI
def extract_keywords_ai(text, num_keywords=10):
    chat_messages = [
        SystemMessage(content='You are an expert assistant with expertise in extracting keywords from text.'),
        HumanMessage(content=f'Please extract the top {num_keywords} keywords from the following text:\nTEXT: {text}')
    ]
    response = llm(chat_messages).content
    keywords = response.split(',')
    return [keyword.strip() for keyword in keywords]

# Function to create a word cloud
def create_word_cloud(text):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

    # Create a matplotlib figure
    fig, ax = plt.subplots()
    ax.imshow(wordcloud, interpolation='bilinear')
    ax.axis('off')

    return fig

# Function for the keyword extractor and word cloud page
def keyword_extractor_page():
    st.image("/content/drive/MyDrive/TM Working/Simplifying/word cloud.jpg", use_column_width=True)
    #st.title("Keyword Extractor and Word Cloud Generator")
    text = st.text_area("Enter the text:")

    if st.button("Extract Keywords and Generate Word Cloud"):
        if text:
            keywords = extract_keywords_ai(text)
            st.write("Keywords:")
            for keyword in keywords:
                st.write(keyword)

            st.write("Word Cloud:")
            fig = create_word_cloud(text)
            st.pyplot(fig)
        else:
            st.write("Please enter the text to analyze.")

# ---------------------------------------------------------------------------------------

def main():
    st.sidebar.title("SummarAIze")
    pages = ["Home", "About", "Need for Study","Libraries Used","Pros and Cons", "API","Summarize Text", "Ask about PDF","Word Cloud","Conclusion","References"]
    choice = st.sidebar.radio("Go to", pages)

    if choice == "Home":
        home_page()
    elif choice == "About":
        about_page()
    elif choice == "Need for Study":
        insights_page()
    elif choice == "Libraries Used":
        lor()
    elif choice == "Pros and Cons":
        pros()
    elif choice == "Conclusion":
        conclude()
    elif choice == "References":
        references()
    elif choice == "Summarize Text":
        text_summarizer_page()
    elif choice == "Ask about PDF":
        pdf_summarizer_page()
    elif choice == "API":
        apikey()
    elif choice == "Word Cloud":
        keyword_extractor_page()

    # Add author names at the end of the sidebar
    st.sidebar.markdown("__________")
    st.sidebar.markdown("Project By")
    st.sidebar.markdown("Mushtari Khan :-) ")
    st.sidebar.markdown("Yuva Priyanka Manda :-)")

# Run the main function
if __name__ == "__main__":
    main()

```
